```python
#importing Libaries and dataset
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

train_df = pd.read_csv('E:\\Laptop\\Clickstream\\Data\\test_data - test_data.csv')
test_df = pd.read_csv('E:\\Laptop\\Clickstream\\Data\\train_data - train_data.csv')
```

```python
#Country vs Purchase

country_conversion = pd.crosstab(
    train_df["country"],
    train_df["price_2"],
    normalize="index"
) * 100

pd.set_option("display.max_rows", None)

country_conversion
```

```python
#Visualization of country vs purchase conversion
country_conversion[1].sort_values(
    ascending=False
).head(15).plot(
    kind="bar",
    figsize=(12,5)
)

plt.ylabel("Purchase %")
plt.title("Top Countries by Purchase Rate")

plt.show()
```

```python
#Category vs Purchase

category_conversion = pd.crosstab(
    train_df["page1_main_category"],
    train_df["price_2"],
    normalize="index"
) * 100

category_conversion
```

```python
#Average Price by Category

train_df.groupby(
    "page1_main_category"
)["price"].mean().sort_values(
    ascending=False
)
```

```python
train_df.groupby(
    "page1_main_category"
)["price"].mean().plot(
    kind="bar"
)

plt.title("Average Price by Category")
plt.show()
```

```python
#Average Price by Country

country_price = train_df.groupby(
    "country"
)["price"].mean()

country_price.sort_values(
    ascending=False)
```

```python
#Numerical Correlation

num_cols = [
    "month",
    "day",
    "order",
    "price"
]

corr = train_df[num_cols].corr()

plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.show()
```
