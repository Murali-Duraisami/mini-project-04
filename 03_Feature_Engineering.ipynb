```python
#Top Countries

top_countries = (
    train_df_fe["country"]
    .value_counts()
    .head(9)
    .index
)
top_countries
```

```python
#Country feature Grouping

import numpy as np

train_df_fe["country_grouped"] = np.where(
    train_df_fe["country"].isin(top_countries),
    train_df_fe["country"].astype(str),
    "Other"
)

train_df_fe["country_grouped"].value_counts()
```

```python
#Colour feature grouping

top_colours = (
    train_df_fe["colour"]
    .value_counts()
    .head(9)
    .index
)


train_df_fe["colour_grouped"] = np.where(
    train_df_fe["colour"].isin(top_colours),
    train_df_fe["colour"].astype(str),
    "Other"
)

train_df_fe["colour_grouped"].value_counts()
```

```python
#Dropping original features 

train_df_fe = train_df_fe.drop(
    columns=[
        "year",
        "session_id",
        "country",
        "colour"
        # "page2_clothing_model"
    ]
)
```

```python
#create function for day_grouping

def day_bucket(x):
    if x <= 10:
        return "start_month"
    elif x <= 20:
        return "mid_month"
    else:
        return "end_month"
```

```python
#Apply day_grouping to the day feature

train_df_fe["day_grouped"] = train_df_fe["day"].apply(day_bucket)
```
